```python
def compute_discrete_ft(eeg, window_length, sampling_rate):
    w_size = window_length * sampling_rate
    nyquist = (w_size // 2) + 1
    n_epochs = len(eeg) // w_size
    
    fourier = np.zeros((n_epochs, nyquist), dtype=complex)
    n = np.arange(w_size)  # Sample indices 0, 1, ..., N-1

    for e in range(n_epochs):
        signal = eeg[e * w_size:(e + 1) * w_size]

        for k in range(nyquist):
            basis = np.exp(-1j * 2 * np.pi * k * n / w_size)  # Complex exponential
            fourier[e][k] = np.sum(signal * basis) / w_size    # Normalize by N

    return fourier
```
